# Logistic regression using Perceptron Trick algorithm

## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generating data

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

## Ploting the points

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

## Perceptron trick algorithm

```python
X_new = np.insert(X,0,np.ones(X.shape[0]),1)

epoch = 1000
lr = 0.01
w = np.ones(X_new.shape[1])

for itr in range(epoch):
    n = np.random.randint(0,X_new.shape[0])
    x_rnd = X_new[n]
    y_hat = np.heaviside(np.dot(w,x_rnd.transpose()),0)
    w = w + lr * (y[n] - y_hat) * x_rnd
```

## Ploting the classification line

```python
x_plot = np.linspace(-2,2,num=50)
y_plot = -(w[0]+w[1]*x_plot)/w[2]
plt.plot(x_plot,y_plot,'-',linewidth=2)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(2,-3)
```
